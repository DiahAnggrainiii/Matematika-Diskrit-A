# 2. Tabel pivot konsonan dan vokal dalam kata

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re

def analisis_per_bagian(path_txt, judul_bagian):
    with open(path_txt, "r", encoding="utf-8") as f:
        text = f.read()

    words = re.findall(r"[A-Za-z√Ä-√ñ√ò-√∂√∏-√ø]+", text)
    words = [w.lower() for w in words if w.isalpha()]

    vowels = set("aiueo")
    data = []
    for w in words:
        v = sum(1 for c in w if c in vowels)
        k = sum(1 for c in w if c.isalpha() and c not in vowels)
        data.append((w, k, v))

    df = pd.DataFrame(data, columns=["Kata", "Konsonan", "Vokal"])

    pivot_counts = df.pivot_table(index="Vokal", columns="Konsonan",
                                  values="Kata", aggfunc="count", fill_value=0)

    pivot_counts = pivot_counts.sort_index(axis=0).sort_index(axis=1)

    pivot_counts.loc["Total"] = pivot_counts.sum(axis=0)
    pivot_counts["Total"] = pivot_counts.sum(axis=1)

    total_kata = df.shape[0]
    pivot_percent = (pivot_counts / total_kata * 100).round(2)

    print(f"\nüìò {judul_bagian}\n")
    print("Tabel 1: Jumlah Kata berdasarkan Jumlah Konsonan dan Vokal")
    print("Jumlah Konsonan ‚Üí")
    print("Jumlah Vokal ‚Üì")
    display(pivot_counts.astype(int))

    print("\nTabel 2: Persentase (%) Kombinasi Konsonan‚ÄìVokal")
    print("Jumlah Konsonan ‚Üí")
    print("Jumlah Vokal ‚Üì")
    display(pivot_percent)

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    cmap = "YlGnBu"

    counts_core  = pivot_counts.iloc[:-1, :-1]
    percent_core = pivot_percent.iloc[:-1, :-1]

    ax = axes[0]
    sns.heatmap(counts_core, annot=True, fmt="d", cmap=cmap,
                cbar=True, ax=ax, linewidths=0.5, vmin=0)
    ax.invert_yaxis()
    ax.set_title(f"Heatmap Cacah Kata ‚Äî {judul_bagian}\n")
    ax.set_xlabel("Jumlah Konsonan")
    ax.set_ylabel("Jumlah Vokal")

    ax2 = axes[1]
    sns.heatmap(percent_core, annot=True, fmt=".2f", cmap=cmap,
                cbar=True, ax=ax2, linewidths=0.5, vmin=0)
    ax2.invert_yaxis()
    ax2.set_title(f"Heatmap Persentase (%) ‚Äî {judul_bagian}\n")
    ax2.set_xlabel("Jumlah Konsonan")
    ax2.set_ylabel("Jumlah Vokal")

    plt.tight_layout()
    plt.show()

analisis_per_bagian("abstrak.txt", "ABSTRAK")
analisis_per_bagian("pendahuluan.txt", "PENDAHULUAN")
analisis_per_bagian("kesimpulan.txt", "KESIMPULAN")
```

Pembahasan :

Berdasarkan hasil tabel dan heatmap, dapat dilihat bahwa sebagian besar kata dalam artikel memiliki 3‚Äì5 huruf konsonan dan 2‚Äì4 huruf vokal. Kombinasi ini muncul paling sering, terlihat dari area berwarna biru tua pada heatmap yang menunjukkan nilai tertinggi. Pola ini menunjukkan bahwa kata-kata dalam teks umumnya memiliki panjang sedang dan struktur yang seimbang antara konsonan dan vokal, sehingga mudah dibaca dan terdengar alami.

Sementara itu, kata dengan jumlah konsonan atau vokal yang terlalu sedikit maupun terlalu banyak jarang muncul, yang ditunjukkan dengan warna yang lebih terang. Hal ini menandakan bahwa penulis lebih banyak menggunakan kata yang umum dalam bahasa Indonesia formal, bukan kata yang terlalu sederhana atau terlalu kompleks.

Secara keseluruhan, hasil ini menunjukkan bahwa teks memiliki gaya penulisan yang rapi dan seimbang, dengan penggunaan kata yang efisien dan mudah dipahami pembaca.

# 3. Jumlah kata dalam kalimat

```python
from matplotlib.colors import LinearSegmentedColormap

def pecah_kalimat_id(text: str):
    text = re.sub(r'\s+', ' ', text).strip()
    potong = re.split(r'(?<=[.!?])\s+', text)
    kalimat = [k.strip() for k in potong if k.strip()]
    return kalimat

def hitung_kata(k: str):
    return len(re.findall(r"[A-Za-z√Ä-√ñ√ò-√∂√∏-√ø]+", k))

def histogram_jumlah_kata_per_kalimat(path_txt: str, judul_bagian: str):
    with open(path_txt, "r", encoding="utf-8") as f:
        text = f.read()

    kalimat = pecah_kalimat_id(text)
    jumlah_kata = [hitung_kata(k) for k in kalimat]

    if len(jumlah_kata) == 0:
        print(f"\nüìò {judul_bagian.upper()}\nTidak ada kalimat terdeteksi.")
        return

    df = pd.DataFrame({
        "Kalimat ke-": np.arange(1, len(kalimat)+1),
        "Jumlah Kata": jumlah_kata
    })

    total_row = pd.DataFrame({
        "Kalimat ke-": ["Total"],
        "Jumlah Kata": [sum(jumlah_kata)]
    })
    df_total = pd.concat([df, total_row], ignore_index=True)

    arr = np.array(jumlah_kata)
    mean, median = arr.mean(), np.median(arr)
    total_kata = int(arr.sum())

    print(f"\nüìò {judul_bagian.upper()}")
    print(f"Total kalimat : {len(arr)}")
    print(f"Total kata    : {total_kata} kata")
    print(f"Terpendek     : {arr.min()} kata")
    print(f"Terpanjang    : {arr.max()} kata")
    print(f"Rata-rata     : {mean:.2f} kata")
    print(f"Median        : {median:.2f} kata\n")

    print("üìä Jumlah kata pada setiap kalimat:")
    display(df_total)

    warna_baru = LinearSegmentedColormap.from_list(
        "DibbySoft", ["#A8E6CF", "#FFD3B6", "#FFAAA5", "#D8B7FF"]
    )

    plt.figure(figsize=(9, 4))
    n, bins, patches = plt.hist(arr, bins='auto', edgecolor="black", linewidth=0.8)
    norm = plt.Normalize(n.min(), n.max())
    for c, p in zip(n, patches):
        plt.setp(p, "facecolor", warna_baru(norm(c)))

    plt.axvline(mean, color="#6A0DAD", linestyle="--", linewidth=1.5,
                label=f"Rata-rata ‚âà {mean:.1f}")

    plt.text(mean + 0.3, max(n)*0.9, f"Rata-rata: {mean:.2f}",
             color="#4A148C", fontsize=10, weight="bold")

    plt.title(f"DISTRIBUSI JUMLAH KATA PER KALIMAT ‚Äî {judul_bagian.upper()}",
              fontsize=12, fontweight="bold", color="#4A148C")
    plt.xlabel("Jumlah Kata per Kalimat")
    plt.ylabel("Frekuensi Kalimat")
    plt.grid(axis='y', linestyle="--", alpha=0.4)
    plt.legend()
    plt.tight_layout()
    plt.show()

histogram_jumlah_kata_per_kalimat("abstrak.txt", "ABSTRAK")
histogram_jumlah_kata_per_kalimat("pendahuluan.txt", "PENDAHULUAN")
histogram_jumlah_kata_per_kalimat("kesimpulan.txt", "KESIMPULAN")
```

Penjelasan :

Berdasarkan hasil analisis jumlah kata per kalimat, setiap bagian teks memiliki panjang kalimat yang berbeda-beda.

Pada bagian abstrak, terdapat 8 kalimat dengan panjang antara 11‚Äì25 kata, dan rata-rata 19 kata per kalimat. Hal ini menunjukkan bahwa abstrak ditulis secara padat dan konsisten, langsung menjelaskan inti penelitian tanpa kalimat berlebihan.

Bagian pendahuluan memiliki 28 kalimat dengan panjang antara 8‚Äì49 kata, dan rata-rata 22 kata per kalimat. Panjang kalimat yang lebih besar ini wajar karena pendahuluan berisi uraian teori, latar belakang, serta konteks penelitian, sehingga kalimatnya lebih panjang dan kompleks.

Sementara itu, bagian kesimpulan terdiri dari 12 kalimat dengan panjang antara 10‚Äì32 kata, dan rata-rata 19 kata per kalimat. Kalimat di bagian ini tergolong sedang, karena berfokus menyampaikan hasil penelitian secara ringkas dan jelas.

Dari bentuk histogram, terlihat bahwa sebagian besar kalimat pada ketiga bagian berada di rentang panjang sedang, sedangkan kalimat yang terlalu pendek atau terlalu panjang jumlahnya sedikit.
Secara keseluruhan, penulisan pada setiap bagian sudah seimbang dan mudah dipahami, menunjukkan gaya bahasa yang efektif untuk menyampaikan isi penelitian.

# 4) Jumlah kalimat dalam paragraf

```python
def load_and_normalize_text(path: str) -> str:
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()
    text = text.replace('\r\n', '\n').replace('\r', '\n')
    lines = [ln.lstrip() for ln in text.split('\n')]
    text = '\n'.join(lines)
    text = re.sub(r'\n{3,}', '\n\n', text).strip()
    return text

def split_paragraphs_strict(text: str):
    paras = re.split(r'\n\s*\n', text)
    paras = [re.sub(r'\s*\n\s*', ' ', p).strip() for p in paras if p.strip()]
    return paras

def split_sentences_id(paragraph: str):
    sents = re.split(r'(?<=[.!?])\s+', paragraph.strip())
    return [s.strip() for s in sents if s.strip()]

def analisis_paragraf(path_txt, judul_bagian):
    text  = load_and_normalize_text(path_txt)
    paras = split_paragraphs_strict(text)

    jumlah_kalimat = [len(split_sentences_id(p)) for p in paras]
    if not jumlah_kalimat:
        print(f"\n‚ö† Tidak ada paragraf/kalimat terbaca di {judul_bagian.upper()}")
        return

    arr = np.array(jumlah_kalimat, dtype=int)

    print(f"\nüìò {judul_bagian.upper()}")
    print(f"Total paragraf : {len(paras)}")
    print(f"Total kalimat  : {arr.sum()}")
    print(f"Terpendek      : {arr.min()} kalimat/paragraf")
    print(f"Terpanjang     : {arr.max()} kalimat/paragraf")
    print(f"Rata-rata      : {arr.mean():.2f} kalimat/paragraf")
    print(f"Median         : {np.median(arr):.2f} kalimat/paragraf\n")

    df = pd.DataFrame({"Paragraf ke-": np.arange(1, len(paras)+1),
                       "Jumlah Kalimat": arr})
    df_total = pd.concat([
        df,
        pd.DataFrame({"Paragraf ke-": ["Total"], "Jumlah Kalimat": [int(arr.sum())]})
    ], ignore_index=True)

    print("üìä Jumlah kalimat dalam tiap paragraf:")
    display(df_total)

    warna_baru = LinearSegmentedColormap.from_list(
        "DibbySoft", ["#A8E6CF", "#FFD3B6", "#FFAAA5", "#D8B7FF"]
    )

    plt.figure(figsize=(9, 4))
    bins = np.arange(1, arr.max() + 2)
    n, bins, patches = plt.hist(arr, bins=bins, edgecolor="black", linewidth=0.8)

    norm = plt.Normalize(n.min() if n.size else 0, n.max() if n.size else 1)
    for c, p in zip(n, patches):
        p.set_facecolor(warna_baru(norm(c)))

    mean = arr.mean()
    plt.axvline(mean, color="#6A0DAD", linestyle="--", linewidth=1.5,
                label=f"Rata-rata ‚âà {mean:.1f}")

    plt.title(f"DISTRIBUSI JUMLAH KALIMAT PER PARAGRAF ‚Äî {judul_bagian.upper()}",
              fontsize=12, fontweight="bold", color="#4A148C")
    plt.xlabel("Jumlah Kalimat per Paragraf")
    plt.ylabel("Frekuensi Paragraf")
    plt.grid(axis='y', linestyle="--", alpha=0.4)
    plt.legend()
    plt.tight_layout()
    plt.show()

analisis_paragraf("abstrak.txt", "ABSTRAK")
analisis_paragraf("pendahuluan.txt", "PENDAHULUAN")
analisis_paragraf("kesimpulan.txt", "KESIMPULAN")
```

Penjelasan :

Berdasarkan hasil analisis jumlah kalimat per paragraf, menunjukkan bahwa setiap bagian memiliki jumlah kalimat per paragraf yang berbeda. Pada bagian Abstrak, terdapat 1 paragraf yang berisi 8 kalimat, menunjukkan bahwa bagian ini ditulis padat dan langsung menjelaskan inti penelitian. Pada bagian Pendahuluan, jumlah paragraf lebih banyak dengan total 28 kalimat yang tersebar dalam beberapa paragraf. Hal ini menunjukkan bahwa penulis menyampaikan uraian latar belakang dan konteks penelitian secara panjang dan terperinci. Sedangkan bagian Kesimpulan terdiri dari beberapa paragraf dengan total 12 kalimat, menandakan bahwa bagian ini disusun lebih singkat untuk menegaskan hasil dan makna dari penelitian. Secara keseluruhan, bagian pendahuluan memiliki kalimat paling banyak dalam tiap paragraf, sementara abstrak dan kesimpulan lebih ringkas sesuai dengan fungsi masing-masing dalam struktur artikel ilmiah.

Berdasarkan bentuk histogram, terlihat bahwa distribusi jumlah kalimat per paragraf meningkat tajam pada bagian pendahuluan dibandingkan dua bagian lainnya. Pola ini menunjukkan bahwa penulis memberikan porsi penjelasan yang lebih besar pada pendahuluan, sedangkan abstrak dan kesimpulan disusun lebih ringkas sesuai fungsinya dalam artikel ilmiah.

```python
dibbysoft = LinearSegmentedColormap.from_list(
    "DibbySoftPlus", ["#A8E6CF", "#FFD3B6", "#FFAAA5", "#D8B7FF"]
)
pal = {"ABSTRAK":"#A8E6CF","PENDAHULUAN":"#FFD3B6","KESIMPULAN":"#D8B7FF"}
order = ["ABSTRAK","PENDAHULUAN","KESIMPULAN"]
np.random.seed(42)

def words_only(s: str):
    return re.findall(r"[A-Za-z√Ä-√ñ√ò-√∂√∏-√ø]+", s)

def split_sentences(text: str):
    s = re.split(r'(?<=[.!?])\s+', re.sub(r'\s+', ' ', text).strip())
    return [x.strip() for x in s if x.strip()]

def normalize_text_file(path: str) -> str:
    with open(path, "r", encoding="utf-8") as f:
        t = f.read()
    t = t.replace("\r\n","\n").replace("\r","\n")
    t = "\n".join([ln.lstrip() for ln in t.split("\n")])
    t = re.sub(r'\n{3,}', '\n\n', t).strip()
    return t

def split_paragraphs(text: str):
    paras = re.split(r'\n\s*\n', text)
    paras = [re.sub(r'\s*\n\s*', ' ', p).strip() for p in paras if p.strip()]
    return paras

def boxplot_cv_per_word():
    rows = []
    for path, label in [("abstrak.txt","ABSTRAK"),
                        ("pendahuluan.txt","PENDAHULUAN"),
                        ("kesimpulan.txt","KESIMPULAN")]:
        txt = normalize_text_file(path)
        toks = [w.lower() for w in words_only(txt)]
        vowels = set("aiueo")
        for w in toks:
            v = sum(1 for c in w if c in vowels)
            k = sum(1 for c in w if c.isalpha() and c not in vowels)
            rows.append((label, v, k))
    if not rows:
        print("‚ö† Tidak ada kata terbaca."); return
    df = pd.DataFrame(rows, columns=["Bagian","Vokal","Konsonan"])

    fig, axes = plt.subplots(1, 2, figsize=(11, 4), sharey=False)

    sns.boxplot(data=df, x="Bagian", y="Vokal", order=order,
                palette=pal, showmeans=True, showfliers=False,
                meanprops={"marker":"o","markerfacecolor":"black","markeredgecolor":"black","markersize":5},
                medianprops={"color":"#4A148C","linewidth":2},
                whiskerprops={"color":"#4A148C","linewidth":1},
                capprops={"color":"#4A148C","linewidth":1},
                boxprops={"edgecolor":"#4A148C","linewidth":1.2}, ax=axes[0])
    sns.stripplot(data=df, x="Bagian", y="Vokal", order=order, hue="Bagian",
                  palette=pal, dodge=False, alpha=0.45, size=3.5,
                  edgecolor="black", linewidth=0.25, legend=False, jitter=0.25, ax=axes[0])
    axes[0].set_title("Sebaran Jumlah Vokal per Kata", color="#4A148C", fontweight="bold")
    axes[0].set_xlabel(""); axes[0].set_ylabel("Frekuensi Vokal / Kata"); axes[0].grid(axis='y', ls='--', alpha=0.35)
    axes[0].set_facecolor("#F7F9FC")

    sns.boxplot(data=df, x="Bagian", y="Konsonan", order=order,
                palette=pal, showmeans=True, showfliers=False,
                meanprops={"marker":"o","markerfacecolor":"black","markeredgecolor":"black","markersize":5},
                medianprops={"color":"#4A148C","linewidth":2},
                whiskerprops={"color":"#4A148C","linewidth":1},
                capprops={"color":"#4A148C","linewidth":1},
                boxprops={"edgecolor":"#4A148C","linewidth":1.2}, ax=axes[1])
    sns.stripplot(data=df, x="Bagian", y="Konsonan", order=order, hue="Bagian",
                  palette=pal, dodge=False, alpha=0.45, size=3.5,
                  edgecolor="black", linewidth=0.25, legend=False, jitter=0.25, ax=axes[1])
    axes[1].set_title("Sebaran Jumlah Konsonan per Kata", color="#4A148C", fontweight="bold")
    axes[1].set_xlabel(""); axes[1].set_ylabel("Frekuensi Konsonan / Kata"); axes[1].grid(axis='y', ls='--', alpha=0.35)
    axes[1].set_facecolor("#F7F9FC")

    plt.tight_layout(); plt.show()


def boxplot_words_per_sentence():
    def load_len(path, label):
        t = normalize_text_file(path)
        lens = [len(words_only(s)) for s in split_sentences(t)]
        return pd.DataFrame({"Bagian": label, "Jumlah Kata per Kalimat": lens})
    df = pd.concat([
        load_len("abstrak.txt","ABSTRAK"),
        load_len("pendahuluan.txt","PENDAHULUAN"),
        load_len("kesimpulan.txt","KESIMPULAN")
    ], ignore_index=True)

    plt.figure(figsize=(9,4))
    sns.boxplot(data=df, x="Bagian", y="Jumlah Kata per Kalimat", order=order,
                palette=pal, showmeans=True, showfliers=False,
                meanprops={"marker":"o","markerfacecolor":"black","markeredgecolor":"black","markersize":5},
                medianprops={"color":"#4A148C","linewidth":2},
                whiskerprops={"color":"#4A148C","linewidth":1},
                capprops={"color":"#4A148C","linewidth":1},
                boxprops={"edgecolor":"#4A148C","linewidth":1.2})
    sns.stripplot(data=df, x="Bagian", y="Jumlah Kata per Kalimat", order=order,
                  hue="Bagian", palette=pal, dodge=False, alpha=0.5,
                  size=5, edgecolor="black", linewidth=0.3, legend=False, jitter=0.25)
    plt.title("Sebaran Jumlah Kata per Kalimat", color="#4A148C", fontweight="bold")
    plt.xlabel(""); plt.ylabel("Frekuensi Kata / Kalimat"); plt.grid(axis='y', ls='--', alpha=0.35)
    plt.gca().set_facecolor("#F7F9FC")
    plt.tight_layout(); plt.show()


def boxplot_sentences_per_paragraph():
    def load_cnt(path, label):
        t = normalize_text_file(path)
        paras = split_paragraphs(t)
        lens = [len(split_sentences(p)) for p in paras]
        return pd.DataFrame({"Bagian": label, "Jumlah Kalimat per Paragraf": lens})
    df = pd.concat([
        load_cnt("abstrak.txt","ABSTRAK"),
        load_cnt("pendahuluan.txt","PENDAHULUAN"),
        load_cnt("kesimpulan.txt","KESIMPULAN")
    ], ignore_index=True)

    plt.figure(figsize=(9,4))
    sns.boxplot(data=df, x="Bagian", y="Jumlah Kalimat per Paragraf", order=order,
                palette=pal, showmeans=True, showfliers=False,
                meanprops={"marker":"o","markerfacecolor":"black","markeredgecolor":"black","markersize":5},
                medianprops={"color":"#4A148C","linewidth":2},
                whiskerprops={"color":"#4A148C","linewidth":1},
                capprops={"color":"#4A148C","linewidth":1},
                boxprops={"edgecolor":"#4A148C","linewidth":1.2})
    sns.stripplot(data=df, x="Bagian", y="Jumlah Kalimat per Paragraf", order=order,
                  hue="Bagian", palette=pal, dodge=False, alpha=0.5,
                  size=5, edgecolor="black", linewidth=0.3, legend=False, jitter=0.25)
    plt.title("Sebaran Jumlah Kalimat per Paragraf", color="#4A148C", fontweight="bold")
    plt.xlabel(""); plt.ylabel("Frekuensi Kalimat / Paragraf"); plt.grid(axis='y', ls='--', alpha=0.35)
    plt.gca().set_facecolor("#F7F9FC")
    plt.tight_layout(); plt.show()

boxplot_cv_per_word()
boxplot_words_per_sentence()
boxplot_sentences_per_paragraph()
```

Penjelasan :

Boxplot berfungsi untuk menampilkan persebaran data melalui nilai minimum, maksimum, median, dan rentang antar kuartil.

Pada konsonan dan vokal per kata, boxplot menunjukkan sebaran jumlah huruf konsonan dan vokal dalam tiap kata. Median konsonan lebih tinggi dibanding vokal di semua bagian, menandakan struktur kata cenderung padat konsonan. Sebaran pada pendahuluan lebih lebar, menunjukkan variasi panjang kata yang lebih besar.

Pada jumlah kata per kalimat, boxplot menunjukkan sebaran panjang kalimat. Pendahuluan memiliki median dan rentang paling tinggi, artinya kalimatnya lebih panjang dan bervariasi. Abstrak dan kesimpulan tampak memiliki sebaran lebih sempit, menandakan kalimatnya lebih singkat dan seragam.

Pada jumlah kalimat per paragraf, boxplot menunjukkan distribusi banyaknya kalimat di tiap paragraf. Pendahuluan memiliki sebaran lebih luas dengan median lebih tinggi, menunjukkan paragraf lebih kompleks. Sementara abstrak dan kesimpulan memiliki median rendah, menunjukkan struktur paragraf yang padat dan ringkas.

# 2. ECDF

```python
def words_only(s: str):
    return re.findall(r"[A-Za-z√Ä-√ñ√ò-√∂√∏-√ø]+", s)

def sentences(text: str):
    s = re.split(r'(?<=[.!?])\s+', re.sub(r'\s+', ' ', text).strip())
    return [x.strip() for x in s if x.strip()]

def normalize_text_file(path: str) -> str:
    with open(path, "r", encoding="utf-8") as f:
        t = f.read()
    t = t.replace("\r\n","\n").replace("\r","\n")
    t = "\n".join([ln.lstrip() for ln in t.split("\n")])
    t = re.sub(r'\n{3,}', '\n\n', t).strip()
    return t

def split_paragraphs(text: str):
    paras = re.split(r'\n\s*\n', text)
    paras = [re.sub(r'\s*\n\s*', ' ', p).strip() for p in paras if p.strip()]
    return paras

def ecdf(arr):
    x = np.sort(arr)
    y = np.arange(1, len(x)+1)/len(x) if len(x) else np.array([])
    return x, y

palette = {"ABSTRAK":"#5AC8FA", "PENDAHULUAN":"#50E3C2", "KESIMPULAN":"#BD93F9"}


def ecdf_cv_per_word():
    data_v = {}
    data_k = {}
    vowels = set("aiueo")

    for label, path in [("ABSTRAK","abstrak.txt"),
                        ("PENDAHULUAN","pendahuluan.txt"),
                        ("KESIMPULAN","kesimpulan.txt")]:
        txt = normalize_text_file(path)
        toks = [w.lower() for w in words_only(txt)]
        vok = [sum(1 for c in w if c in vowels) for w in toks]
        kon = [sum(1 for c in w if c.isalpha() and c not in vowels) for w in toks]
        data_v[label] = np.array(vok, dtype=int)
        data_k[label] = np.array(kon, dtype=int)

    fig, axes = plt.subplots(1, 2, figsize=(12,4))

    for label in ["ABSTRAK","PENDAHULUAN","KESIMPULAN"]:
        x, y = ecdf(data_v[label])
        if len(x): axes[0].step(x, y, where='post', label=label, lw=2, color=palette[label])
    axes[0].set_xlabel("Jumlah Vokal per Kata")
    axes[0].set_ylabel("Proporsi ‚â§ x (ECDF)")
    axes[0].set_title("ECDF ‚Äî Vokal per Kata (makin kiri = makin vokal sedikit)", fontweight="bold")
    axes[0].grid(ls='--', alpha=0.4)
    axes[0].legend()

    for label in ["ABSTRAK","PENDAHULUAN","KESIMPULAN"]:
        x, y = ecdf(data_k[label])
        if len(x): axes[1].step(x, y, where='post', label=label, lw=2, color=palette[label])
    axes[1].set_xlabel("Jumlah Konsonan per Kata")
    axes[1].set_ylabel("Proporsi ‚â§ x (ECDF)")
    axes[1].set_title("ECDF ‚Äî Konsonan per Kata (makin kiri = makin konsonan sedikit)", fontweight="bold")
    axes[1].grid(ls='--', alpha=0.4)
    axes[1].legend()

    plt.tight_layout(); plt.show()

def ecdf_words_per_sentence():
    plt.figure(figsize=(9,4))
    for label, path in [("ABSTRAK","abstrak.txt"),
                        ("PENDAHULUAN","pendahuluan.txt"),
                        ("KESIMPULAN","kesimpulan.txt")]:
        txt = normalize_text_file(path)
        lens = np.array([len(words_only(s)) for s in sentences(txt)], dtype=int)
        x, y = ecdf(lens)
        if len(x): plt.step(x, y, where='post', label=label, lw=2, color=palette[label])

    plt.xlabel("Jumlah Kata per Kalimat")
    plt.ylabel("Proporsi ‚â§ x (ECDF)")
    plt.title("ECDF ‚Äî Panjang Kalimat (makin kiri = makin ringkas)", fontweight="bold")
    plt.grid(ls='--', alpha=0.4)
    plt.legend()
    plt.tight_layout(); plt.show()

def ecdf_sentences_per_paragraph():
    plt.figure(figsize=(9,4))
    for label, path in [("ABSTRAK","abstrak.txt"),
                        ("PENDAHULUAN","pendahuluan.txt"),
                        ("KESIMPULAN","kesimpulan.txt")]:
        txt = normalize_text_file(path)
        paras = split_paragraphs(txt)
        lens = np.array([len(sentences(p)) for p in paras], dtype=int)
        x, y = ecdf(lens)
        if len(x): plt.step(x, y, where='post', label=label, lw=2, color=palette[label])

    plt.xlabel("Jumlah Kalimat per Paragraf")
    plt.ylabel("Proporsi ‚â§ x (ECDF)")
    plt.title("ECDF ‚Äî Kepadatan Paragraf (makin kiri = paragraf lebih singkat)", fontweight="bold")
    plt.grid(ls='--', alpha=0.4)
    plt.legend()
    plt.tight_layout(); plt.show()

ecdf_cv_per_word()
ecdf_words_per_sentence()
ecdf_sentences_per_paragraph()
```

Penjelasan :

ECDF digunakan untuk menunjukkan sebaran kumulatif data, yaitu proporsi nilai yang lebih kecil dari suatu batas tertentu. Semakin cepat kurva naik, semakin banyak data bernilai kecil (lebih ringkas).

Pada konsonan dan vokal per kata, sebagian besar kata memiliki sedikit vokal dan lebih banyak konsonan. Kurva pendahuluan naik lebih lambat, menandakan kata-kata lebih panjang dan bervariasi, sedangkan abstrak dan kesimpulan lebih ringkas.

Pada jumlah kata per kalimat, abstrak dan kesimpulan memiliki kurva yang lebih ke kiri, menunjukkan kalimatnya lebih pendek dan padat. Pendahuluan lebih ke kanan, menandakan kalimatnya lebih panjang dan kompleks.

Pada jumlah kalimat per paragraff, ECDF abstrak tidak tampak jelas karena hanya memiliki satu paragraf, sehingga tidak membentuk kurva. Pendahuluan memiliki kurva yang lebih lebar dan landai karena paragrafnya panjang dan berisi banyak kalimat, sedangkan kesimpulan berada di tengah dengan paragraf yang lebih ringkas.

```python
results = {}
for label, path, color in parts:
    try:
        results[label] = analyze(path)
    except FileNotFoundError:
        print(f"‚ö†Ô∏è File {path} tidak ditemukan.")

print("\n=== (1) Pareto: Vokal & Konsonan per Kata ===")
for label, path, color in parts:
    if label not in results:
        continue
    word_rows, _, _ = results[label]
    pareto_words_stacked(
        word_rows,
        title=f"{label} ‚Äî Vokal & Konsonan per Kata",
        color_base=color,
        topn_labels=12,
        label_every=1,
        rotate_deg=0,
        show_unique=False
    )

print("\n=== (2) Pareto: Kata per Kalimat ===")
for label, path, color in parts:
    if label not in results:
        continue
    _, sent_level, _ = results[label]
    pareto_simple(
        sent_level,
        title=f"{label} ‚Äî Kata per Kalimat",
        xlabel="Kalimat (urut panjang ‚Üí pendek)",
        color=color
    )

print("\n=== (3) Pareto: Kalimat per Paragraf ===")
for label, path, color in parts:
    if label not in results:
        continue
    _, _, para_level = results[label]
    pareto_simple(
        para_level,
        title=f"{label} ‚Äî Kalimat per Paragraf",
        xlabel="Paragraf (urut padat ‚Üí renggang)",
        color=color
    )
```

Penjelasan :

Pareto digunakan untuk menunjukkan bagian elemen yang paling berpengaruh terhadap keseluruhan data. Dalam analisis teks, grafik ini membantu melihat seberapa besar kata, kalimat, dan paragraf berkontribusi pada kepadatan tulisan.

Pada konsonan dan vokal per kata, terlihat hanya sebagian kecil kata dengan jumlah huruf tinggi yang mendominasi total huruf, sedangkan mayoritas kata pendek dan memiliki keseimbangan antara vokal dan konsonan. Tulisan yang tampak menumpuk di bagian bawah grafik terjadi karena banyak kata ditampilkan secara bersamaan pada sumbu X, sehingga label-label kata saling berdekatan dan menimpa satu sama lain. Fenomena ini tidak memengaruhi hasil analisis, hanya merupakan keterbatasan tampilan visual ketika jumlah kata sangat banyak.

Pada jumlah kata per kalimat, mayoritas kalimat memiliki panjang sedang (sekitar 8‚Äì15 kata), menggambarkan penulisan yang padat dan mudah dipahami.

Pada jumlah kalimat per paragraf, sebagian besar paragraf terdiri dari 2‚Äì4 kalimat, menandakan struktur paragraf ringkas dan fokus pada satu pokok pikiran.

# 4. WORD CLOUD

```python
from wordcloud import WordCloud, STOPWORDS

def wordcloud_for(path, title):
    with open(path,'r',encoding='utf-8') as f:
        t = f.read().lower()
    stop_id = set(STOPWORDS)|{"yang","dan","di","ke","untuk","dengan","pada","atau","ini","itu","dalam","sebagai","oleh","karena"}
    wc = WordCloud(width=900, height=400, background_color="white",
                   stopwords=stop_id, collocations=False).generate(t)
    plt.figure(figsize=(9,4))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Word Cloud ‚Äî {title}", fontweight="bold")
    plt.tight_layout()
    plt.show()

wordcloud_for("abstrak.txt", "ABSTRAK")
wordcloud_for("pendahuluan.txt", "PENDAHULUAN")
wordcloud_for("kesimpulan.txt", "KESIMPULAN")
```

Penjelasan :

Word Cloud digunakan untuk menampilkan kata-kata yang paling sering muncul dalam suatu teks. Semakin besar ukuran sebuah kata pada gambar, semakin sering kata tersebut muncul dan semakin besar kontribusinya terhadap isi teks.

Pada bagian Abstrak, Word Cloud menampilkan kata-kata utama yang menggambarkan fokus penelitian secara ringkas.

Pada bagian Pendahuluan, kata yang muncul lebih beragam, menunjukkan konteks dan latar belakang permasalahan yang dibahas.

Sedangkan pada bagian Kesimpulan, kata-kata dominan mencerminkan hasil akhir atau temuan penting dari pembahasan.
