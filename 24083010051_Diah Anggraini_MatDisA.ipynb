# Tahap 5 : Matriks Kovarian dan Matriks Korelasi

Pada tahap kelima ini, akan dihitung nilai matriks kovarian dan matriks korelasi.
Matriks kovarian digunakan untuk memahami hubungan antar variabel dalam satu set data, seperti menunjukkan seberapa kuat dua variabel berubah bersamaan (variabel yang meningkat bersamaan atau menurun secara terbalik). Sedangkan Matriks korelasi digunakan untuk menganalisis dan memvisualisasikan hubungan linear antar variabel dalam sebuah kumpulan data, membantu mengidentifikasi pola, tren, dan kekuatan hubungan tersebut.

```python
kelasku = df[kolom_nilai].to_numpy(dtype=float)

cov_mat = np.cov(kelasku, rowvar=False, ddof=1)
corr_mat = np.corrcoef(kelasku, rowvar=False)

print("Matriks Kovarian:\n", cov_mat, "\n")
print("Matriks Korelasi:\n", corr_mat)
```

Hasil Matriks Kovarian di atas menunjukkan bahwa Matriks kovarian berukuran 4×4 karena menghitung hubungan antara empat variabel, yaitu A, B, C, dan D. Nilai pada diagonal (0.0938, 0.1009, 0.0933, dan 0.1079) menunjukkan varian dari masing-masing variabel, yang berarti keempat variabel memiliki sebaran data yang hampir sama. Nilai di luar diagonal menunjukkan kovarian antar variabel. Kovarian antara A dan B sebesar -0.0113 menandakan hubungan sangat lemah dan berlawanan arah. Kovarian antara A dan C sebesar -0.0089 juga menunjukkan hubungan lemah dan berlawanan arah, sementara antara A dan D sebesar -0.0022 hampir tidak menunjukkan hubungan. Di sisi lain, kovarian antara B dan C sebesar 0.0226, serta antara B dan D sebesar 0.0272, menunjukkan adanya hubungan searah meskipun sangat lemah. Terakhir, kovarian antara C dan D sebesar 0.0129 juga menunjukkan hubungan searah yang lemah.

Secara keseluruhan nilai tersebut dapat disimpulkan bahwa hubungan antar variabel A, B, C, dan D relatif sangat lemah, baik yang searah maupun berlawanan arah.

Hasil Matriks Korelasi diatas, menunjukkan bahwa Matriks korelasi berukuran 4×4 karena menghitung hubungan antara empat variabel, yaitu A, B, C, dan D. Nilai diagonal semuanya 1, karena setiap variabel pasti berkorelasi sempurna dengan dirinya sendiri. Nilai di luar diagonal menunjukkan kekuatan dan arah hubungan antar variabel. Korelasi antara A dan B sebesar -0.116, artinya hubungan negatif sangat lemah, demikian juga antara A dan C sebesar -0.096. Korelasi antara A dan D sebesar -0.022 hampir mendekati nol, yang berarti hampir tidak ada hubungan. Di sisi lain, korelasi antara B dan C sebesar 0.232 menunjukkan hubungan positif lemah, sedangkan B dan D sebesar 0.261 menunjukkan hubungan positif yang juga masih lemah. Terakhir, korelasi antara C dan D sebesar 0.129 juga menunjukkan hubungan positif tetapi lemah.

Secara keseluruhan, matriks korelasi ini memperlihatkan bahwa hubungan antar variabel A, B, C, dan D umumnya lemah, baik yang positif maupun negatif. Tidak ada korelasi yang kuat, sehingga keempat variabel cenderung berdiri sendiri dengan hanya sedikit keterkaitan.

# Tahap 6 : Cosine Similarity

Pada tahap keenam ini, akan dilakukan perhitungan Cosine similarity. Cosine similarity adalah metode untuk mengukur tingkat kemiripan antara dua objek berdasarkan sudut kosinus dari vektor data keduanya. Nilainya berada pada rentang -1 hingga 1, di mana nilai 1 menunjukkan dua objek identik, nilai mendekati 1 berarti sangat mirip, nilai mendekati 0 berarti hampir tidak ada kemiripan, dan nilai negatif berarti berlawanan arah.

```python
from numpy.linalg import norm

X_matrix = df[kolom_nilai].to_numpy(dtype=float)

students = df["NAMA"]

norms = np.linalg.norm(X_matrix, axis=1, keepdims=True)
cosine_sim = (X_matrix @ X_matrix.T) / (norms @ norms.T)

cosine_df = pd.DataFrame(cosine_sim, index=students, columns=students)

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

display(cosine_df)
```

Hasil perhitungan ditampilkan dalam bentuk matriks, di mana setiap baris dan kolom mewakili seorang mahasiswa. Misalnya, cosine similarity antara Madina dan Maria sebesar 0.84 menunjukkan bahwa pola nilai keduanya cukup mirip, atau dengan kata lain sekitar 84% kesamaan. Informasi ini berguna untuk melihat siswa yang memiliki kecenderungan nilai yang serupa pada variabel A, B, C, dan D, sehingga dapat diidentifikasi kelompok siswa dengan karakteristik penilaian yang hampir sama.

# Analisis Tambahan (PCA dan Clustering)

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = df[kolom_nilai].to_numpy(dtype=float)
names = df["NAMA"].astype(str).values

scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

Kode di atas digunakan untuk menyiapkan data sebelum dilakukan analisis PCA dan Clustering. Pertama, nilai dari kolom yang dipilih (kolom_nilai) diambil lalu diubah menjadi matriks numerik. Nama mahasiswa disimpan dalam variabel names agar bisa digunakan kembali pada hasil analisis. Selanjutnya, data distandarisasi menggunakan StandardScaler, yaitu mengubah setiap variabel agar memiliki rata-rata nol dan standar deviasi satu. Proses standarisasi ini penting supaya semua variabel berada pada skala yang sama, sehingga analisis PCA dan K-Means tidak bias terhadap variabel yang memiliki rentang nilai lebih besar.

PCA (Principal Component Analysis)

Digunakan untuk menyederhanakan 4 variabel (A, B, C, D) menjadi 2 komponen utama agar mudah divisualisasikan dan terlihat polanya.

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_std)

print("Rasio varians per komponen:", pca.explained_variance_ratio_)
print("Total varians (PC1+PC2):", pca.explained_variance_ratio_.sum())

loadings = pd.DataFrame(
    pca.components_.T,
    index=kolom_nilai,
    columns=["Loading_PC1", "Loading_PC2"]
)
print("Loading (kontribusi tiap variabel ke komponen):")
display(loadings)

pca_df = pd.DataFrame({
    "NAMA": names,
    "PC1": X_pca[:, 0],
    "PC2": X_pca[:, 1]
})
display(pca_df.head())

plt.figure(figsize=(7,5))
plt.scatter(X_pca[:,0], X_pca[:,1], s=35, alpha=0.8)
plt.title("PCA (2 Komponen) dari Variabel A–D")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True, linestyle="--", alpha=0.3)
plt.show()
```

Hasil PCA menunjukkan bahwa dua komponen utama (PC1 dan PC2) mampu menjelaskan 61,18% variasi data, dengan PC1 menyumbang 36,53% dan PC2 sebesar 24,65%. Artinya, sebagian besar pola dari empat variabel (A: adil, amanah; B: suka membantu; C: nyambung kalau ngobrol; D: asyik, menarik) sudah dapat dirangkum hanya dengan dua komponen ini.

Berdasarkan nilai loading, PC1 terutama dipengaruhi positif oleh B (0,623), C (0,517), dan D (0,511), serta sedikit negatif oleh A (-0,289). Dengan demikian, mahasiswa dengan skor PC1 tinggi cenderung menonjol pada aspek B, C, dan D. Sementara itu, PC2 terutama dipengaruhi oleh A (0,867) dan sebagian oleh D (0,476), sehingga skor PC2 tinggi menggambarkan mahasiswa yang menonjol pada aspek A.

Dari tabel hasil PCA, Maria Dwi Kurniasih memiliki skor PC2 tinggi (1,021) namun PC1 rendah (-0,610), artinya lebih kuat pada aspek A tetapi rendah pada B–C–D. Sebaliknya, Adinda Putri Rachmawati memiliki PC1 positif (0,733) dan PC2 netral, sehingga lebih menonjol pada aspek B–C–D. Contoh lain, Aquina Syabita memiliki PC1 sangat rendah (-2,720) dan PC2 positif (0,322), sehingga lemah pada B–C–D namun cukup tinggi pada A.

Scatter plot memperlihatkan distribusi mahasiswa dalam bidang dua dimensi berdasarkan PC1 dan PC2. Titik-titik yang berdekatan menunjukkan mahasiswa dengan pola nilai serupa, sedangkan titik yang jauh terpisah mencerminkan perbedaan karakteristik yang cukup besar antar mahasiswa.

Clustering (K-Means dengan k=3)

Digunakan untuk mengelompokkan data ke dalam 3 cluster berdasarkan kemiripan nilai pada variabel A, B, C, dan D. Metode ini menghasilkan pembagian kelompok, centroid atau rata-rata nilai tiap cluster, serta visualisasi sebaran data dalam ruang PCA.

```python
k = 3
kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
labels = kmeans.fit_predict(X_std)

df_clusters = df.copy()
df_clusters["Cluster"] = labels
print("Jumlah anggota per cluster:")
display(df_clusters["Cluster"].value_counts().sort_index())

centroids_std = kmeans.cluster_centers_
centroids_original = scaler.inverse_transform(centroids_std)
centroids_df = pd.DataFrame(centroids_original, columns=kolom_nilai)
centroids_df.index = [f"Cluster {i}" for i in range(k)]
print("Centroid tiap cluster (skala asli A–D):")
display(centroids_df)

plt.figure(figsize=(7,5))
for c in range(k):
    idx = labels == c
    plt.scatter(X_pca[idx,0], X_pca[idx,1], s=50, alpha=0.9, label=f"Cluster {c}")
plt.title("K-Means (k=3) divisualisasikan pada ruang PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.3)
plt.show()
```

Hasil clustering dengan K-Means membagi mahasiswa menjadi tiga kelompok. Cluster 0 berjumlah 11 orang dengan karakteristik nilai rata-rata rendah pada aspek adil, amanah (0,20), namun tinggi pada suka membantu (0,79), nyambung kalau ngobrol (0,62), dan asyik, menarik (0,66). Cluster 1 terdiri dari 16 orang dengan nilai rata-rata cukup tinggi pada adil, amanah (0,54), tetapi rendah pada suka membantu (0,24), nyambung kalau ngobrol (0,39), dan asyik, menarik (0,34), sehingga lebih menonjol pada aspek adil, amanah saja. Sementara itu, Cluster 2 berjumlah 13 orang dengan rata-rata nilai tinggi hampir di semua aspek, yaitu adil, amanah (0,85), suka membantu (0,69), nyambung kalau ngobrol (0,69), dan asyik, menarik (0,53). Visualisasi pada ruang PCA menunjukkan pemisahan antar cluster cukup jelas, di mana setiap warna mewakili kelompok dengan pola karakteristik yang mirip.
